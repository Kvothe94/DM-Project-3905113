```python
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import graphviz
import csv
import random
import numpy
import math
from copy import copy, deepcopy
import pandas as pd
random.seed(5)
```

```python
datasets = list()

with open('mushrooms.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    x = list(reader)
    result = numpy.array(x)
    dataset = result[1:,:]
    datasets.append(dataset)

with open('nursery.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    x = list(reader)
    result = numpy.array(x)
    dataset = result[1:,:]
    aux = deepcopy(dataset[:,0])
    dim = len(dataset[0,:])
    l = len(dataset[:,0])
    dataset[:,0] = dataset[:, dim-1]
    for i in range(0,l):
        dataset[i, dim-1] = aux[i]
    datasets.append(dataset)

with open('connect-4.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    x = list(reader)
    result = numpy.array(x)
    dataset = result[1:,:]
    aux = deepcopy(dataset[:,0])
    dim = len(dataset[0,:])
    l = len(dataset[:,0])
    dataset[:,0] = dataset[:, dim-1]
    for i in range(0,l):
        dataset[i, dim-1] = aux[i]
    datasets.append(dataset)
```

```python
for dataset in datasets:
    le = LabelEncoder()
    for i in range(0,len(dataset[0,:])):
        le.fit(dataset[:,i])
        dataset[:,i] = le.transform(dataset[:,i])
    """for row in dataset:
        print(row)"""
```

```python
k = 0
names = ["MUSHROOMS", "NURSERY", "CONNECT-4"]
best_depth_tree = [0,0,0]
for dataset in datasets:
    
    dim_trainingset = int(len(dataset) * 2 / 3)
    shuffled_dataset = deepcopy(dataset)
    random.shuffle(shuffled_dataset)
    training_set = shuffled_dataset[0:dim_trainingset, :]
    test_set = shuffled_dataset[dim_trainingset+1:, :]
    print(names[k])
    print('LENGTH DATASET:', len(dataset), 'LENGTH TRAININGSET:', len(training_set), 'LENGTH TESTSET:', len(test_set),
            '\n')

    max_depths = range(1,50)
    accuracy = [0 for i in max_depths]
    m = 0
    for depth in max_depths:
        
        myDTC = tree.DecisionTreeClassifier(criterion="entropy", max_depth=depth)
        scores = cross_val_score(myDTC, training_set[:,1:], training_set[:,0], cv=5)
        accuracy[m] = scores.mean()
        m = m + 1
        
    plt.plot(max_depths, accuracy)
    plt.ylabel('ACCURACY')
    plt.xlabel('DEPTHS')
    plt.show()
    best_depth_tree[k] = max_depths[numpy.argmax(accuracy)]
    print(best_depth_tree[k])
    k = k + 1
```
